## 了解解析器Xpath
### 概念
Xpath 即为 XML 路径语言（XML Path Language），它是一种用来确定 XML 文档中某部分位置的语言。


### Xpath解析网页流程
1.首先通过Requests库获取网页数据
2.通过网页解析，得到想要的数据或者新的链接


import requests
from lxml import etree 

加请求头headers：
网页右键“检查元素”-Network-Doc 


```python
url = 'https://movie.douban.com/subject/1292052/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
data = requests.get(url, headers=headers).text
s = etree.HTML(data) # 将文本内容转换成一个可被 lxml 模块解析的 HTML 文档对象
print(data) #测试是否能够成功爬取url（需要加上headers信息）
```

```python
# 格式：file = s.xpath('元素的Xpath信息/text()')
# 获取Xpath信息（右键-复制-复制Xpath）
# //*[@id="content"]/h1/span[1]

film = s.xpath('//*[@id="content"]/h1/span[1]/text()')
print(film)
```

```python
# 获取其他信息
director = s.xpath('//*[@id="info"]/span[1]/span[2]/a/text()')
print(director)
```
