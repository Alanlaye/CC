## 3.描述数据

```python
df[['Name of member', 'Amount, £']].groupby("Name of member").agg('sum')
# 从df中选择Name of member和Amount, £这两列，得到一个新的DataFrame
## df[['Name of member', 'Amount, £']]即是这个新的DataFrame

# 按照Name of member列的值进行分组，把相同的名字放一起，得到一个GroupBy对象
## df[['Name of member', 'Amount, £']].groupby("Name of member")
## 查看该GroupBy对象的详细信息（与返回值）：
## df[['Name of member', 'Amount, £']].groupby("Name of member").groups
## >>>{'Adam Afriyie': [0, 1], 'Adam Holloway': [2], ...}

# 最后的 agg(‘sum’) 是用来对每个分组进行聚合操作的
# 可以用其他的聚合函数来替换 ‘sum’
# 比如 ‘mean’ （求平均）、 ‘count’ （计数）、 ‘max’ （求最大值）等
# 也可以用一个列表来传入多个聚合函数
# 比如 agg([‘sum’, ‘mean’, ‘count’]) 
# 这样就会得到一个新的DataFrame对象，其中每一列对应一个聚合函数的结果
```

```python
dat1 = df[['Name of member', 'Amount, £']].groupby("Name of member").agg('sum')
dat1 = dat1.sort_values(by=['Amount, £'], ascending = False)
# 对dat1按照 ‘Amount, £’ 这一列的值进行降序排列（ascending = False）
dat1[:10]
```

```python
dat1['Rank'] = range(1, len(dat1)+1)
# 用 range(1, len(dat1)+1) 生成一个从1到dat1的长度（len(dat1)）的整数序列
# 把这个序列赋值给dat1的 ‘Rank’ 这一列
# 也就是创建一个新的列，名字叫做 ‘Rank’ ，值是名次
```

```python
dat1[:10]
```

```python
plt.figure(figsize =(16, 6), dpi = 100)

plt.plot(dat1['Rank'], dat1['Amount, £'], 'ro')
# 横坐标是dat1的 ‘Rank’ 列,纵坐标是dat1的 ‘Amount, £’ 列
# 点的样式是红色的圆圈（‘ro’）
# plt.yscale('log')
# plt.xscale('log')

plt.xlabel('Rank')
plt.ylabel('Amount, £')
# 设置横坐标和纵坐标的标签

plt.show()
```

```python
plt.figure(figsize =(16, 6), dpi = 100)

dat10 = dat1[:10]
plt.plot(dat10['Rank'], dat10['Amount, £'], 'ro')
# plt.yscale('log')
# plt.xscale('log')

plt.xlabel('Rank')
plt.ylabel('Amount, £')
plt.show()
```

## 4.桑基图

桑基图是一种用线条来表示数据从哪里来到哪里去的图表。线条越粗，表示数据越多。线条的颜色可以表示不同的类别。桑基图可以帮助我们看清数据的变化和分布，比如能源的消耗和转化，或者客户的流失和留存

```python
pip install pyecharts
```

```python
from pyecharts import options as opts # 设置图表样式的选项，比如颜色、字体、标题
from pyecharts.charts import Sankey # 用来创建桑基图对象的类

nodes = [ # 节点名称
    {"name": "category1"},
    {"name": "category2"},
    {"name": "category3"},
    {"name": "category4"},
    {"name": "category5"},
    {"name": "category6"},
]

links = [ # 起点（source）；终点（target）；数据量/连线的宽度（value）
    {"source": "category1", "target": "category2", "value": 10},
    {"source": "category2", "target": "category3", "value": 15},
    {"source": "category3", "target": "category4", "value": 20},
    {"source": "category5", "target": "category6", "value": 25},
]
c = (
    Sankey() # 创建了一个Sankey对象（实例），并赋值给变量c
    .add(
        "sankey", # 定义图标名称
        nodes, # 节点数据：传入前面定义的nodes列表
        links, # 链接数据：传入前面定义的links列表
        linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="source"),
        # 设置链接线条的样式，这里设置了透明度为0.2，曲度为0.5，颜色根据源节点决定
        label_opts=opts.LabelOpts(position="right"),
        # 设置节点标签的样式，这里设置了标签位置在右侧
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Sankey-基本示例"))
    #.render("sankey_base.html")
    .render_notebook()
)

c
```

```python
dat4 = df[['Allowance Type','Expenditure Type', 'Amount, £']].groupby(['Expenditure Type', 'Allowance Type'],  as_index = False).agg('sum')
dat4 = dat4.sort_values(by=['Amount, £'], ascending = False)
dat4
```

```python
nodes = dat4['Allowance Type'].unique().tolist() + dat4['Expenditure Type'].unique().tolist()
nodes = [{'name': i} for i in nodes]
links = [{'source': dat4['Allowance Type'][i], 'target': dat4['Expenditure Type'][i], 'value': dat4['Amount, £'][i]}  for i in dat4.index]
```

```python
from pyecharts import options as opts
from pyecharts.charts import Sankey

s = (
    Sankey()
    .add(
        "sankey",
        nodes,
        links,
        linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="source"),
        label_opts=opts.LabelOpts(position="right"),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Sankey Graph of UK MPS"))
)

s.render_notebook()
```

```python
s.render("sankey_mps.html")
```
